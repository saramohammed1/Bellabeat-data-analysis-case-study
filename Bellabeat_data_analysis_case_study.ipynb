```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/dailyActivity_merged.csv')
df
```

```python
df.Id.nunique()
```

```python
cols =['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes','Calories']
df=df[cols]
df
```

```python
df.rename(columns={'ActivityDate':'date'}, inplace=True)
df
```

```python
df['TotalMinutes']= df.VeryActiveMinutes + df.FairlyActiveMinutes + df.SedentaryMinutes + df.LightlyActiveMinutes
df
```

```python
df['TotalHours'] = round(df.TotalMinutes/60)
df
```

```python
df.info()
```

```python
df.date= pd.to_datetime(df.date)
```

```python
df.info()
```

```python
import datetime as dt
df['DayOfWeek'] = df.date.dt.day_name()
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=((6,4)))
plt.hist(df.DayOfWeek, bins=7, color='lightskyblue',width=0.6)
plt.grid(True)

plt.xlabel('Day of the week')
plt.ylabel('Freq')

plt.title('Freq by Day of Week')
plt.show()
```

```python
sns.heatmap(df.corr())
```

```python
plt.figure(figsize=((8,6)))
plt.scatter(df.TotalSteps, df.Calories, c= df.Calories)

median_steps= 7405
median_Calories= 2134

plt.axhline(median_Calories, color='blue', label='median of calories')
plt.axvline(median_steps, color='red', label='median of steps')

plt.xlabel('Steps')
plt.ylabel('Calories')

plt.title('Calories by Total Steps')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=((8,6)))
plt.scatter(df.TotalHours, df.Calories, c= df.Calories)

median_hours= 24
median_Calories= 2134

plt.axhline(median_Calories, color='blue', label='median of calories')
plt.axvline(median_hours, color='red', label='median of hours')

plt.xlabel('Hours')
plt.ylabel('Calories')

plt.title('Calories by Total Hours')
plt.legend()
plt.show()
```
